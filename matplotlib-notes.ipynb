```python
# Imports:
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### Basic Graph

```python
"""
- Using plt.plot.
- Takes two 1-D arrays, one for x values and one for y values.
- Plots points, taking one co-ordinate from each value, and joins points using straight lines.
- Arrays need to be of same length.
"""

x = [1, 2, 9]
y = [10, 0, 5]

plt.plot(x, y, 
         label = 'random relation', 
         color = 'red', 
         linewidth = 2, linestyle = '--', 
         marker = 'x', markersize = 10, 
         markeredgecolor = 'blue')
"""
- Full list of attributes available in documentation.
- label:
    - Used in relation with legend.
    - Will not be shown unless plt.legend() is called.
- color: accepts hexadecimal values.
- marker: used to add markers where specific points are, similar to 'point' in E.G.
"""

# Title:
plt.title('First ever graph!', fontdict = {'fontname': 'Comic Sans MS', 'fontsize': 20})
"""
- Use fontdict to change style of title font.
"""

# Axis Labels:
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')

# Ticks:
plt.xticks([i for i in range(11)])
plt.yticks([2, 4, 6, 8, 9, 10])
"""
- Ticks, i.e., the values which are displayed on the axes, can be changed manually.
- Possible use case: avoiding decimal values.
- List of ticks needs to be passed.
- Even if difference between ticks is not constant, they will be correctly displayed on the axes.
"""

# Legend:
plt.legend()

plt.show()
```

```python
# Displaying graph correctly:
"""
- The correct way to show graphs is using plt.show()
- This statement CAN be avoided while using jupyter notebooks, due to the way cell outputs work.
- But without plt.show() statement, an extra line is displayed.
"""

x = [1, 2, 9]
y = [10, 0, 5]

plt.plot(x, y)
```

```python
# Shorthand notation for formatting lines:
"""
- Syntax: '<colour><marker><linestyle>'.
- Accepts single letter for colour name.
"""

x = [1, 2, 3]
y = [1, 2, 3]

plt.plot(x, y, 'rX--', label = 'y = x')
plt.legend()
plt.show()
```
