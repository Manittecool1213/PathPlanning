```python
# Imports:
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### Basic Graph

```python
"""
- Using plt.plot.
- Takes two 1-D arrays, one for x values and one for y values.
- Plots points, taking one co-ordinate from each array, and joins these points using straight lines.
- Arrays need to be of same length, otherwise an error is thrown.
"""

x = [1, 2, 9]
y = [10, 0, 5]

plt.plot(x, y, 
         label = 'random relation', 
         color = 'red', 
         linewidth = 2, linestyle = '--', 
         marker = 'x', markersize = 10, 
         markeredgecolor = 'blue')
"""
- Full list of attributes available in documentation.
- label:
    - Used in relation with legend.
    - Will not be shown unless plt.legend() is called.
- color: accepts hexadecimal values.
- marker: used to add markers where specific points are, similar to 'point' in E.G..
"""

# Title:
plt.title('First ever graph!', fontdict = {'fontname': 'Comic Sans MS', 'fontsize': 20})
"""
- Use fontdict to change style of title font.
"""

# Axis Labels:
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')

# Ticks:
plt.xticks([i for i in range(11)])
plt.yticks([2, 4, 6, 8, 9, 10])
"""
- Ticks, i.e., the values which are displayed on the axes, can be changed manually.
- Possible use case: avoiding decimal values for ticks, if plotted points have decimals.
- List of ticks needs to be passed.
- Even if difference between ticks is not constant, they will be correctly displayed on the axes.
"""

# Legend:
plt.legend()
"""
- The legend is automatically moved to a convenient location.
- e.g.] If the graph lies towards the top-right, the legend will be displayed towards the top-left.
"""

plt.show()
```

```python
# Displaying graph correctly:
"""
- The correct way to show graphs is using plt.show()
- This statement CAN be avoided while using jupyter notebooks, due to the way cell outputs work.
- But without plt.show() statement, an extra line is displayed.
"""

x = [1, 2, 9]
y = [10, 0, 5]

plt.plot(x, y)
```

```python
# Shorthand notation for formatting lines:
"""
- Syntax: '<colour><marker><linestyle>'.
- Accepts single letter for colour name.
"""

x = [1, 2, 3]
y = [1, 2, 3]

plt.plot(x, y, 'rX--', label = 'y = x')
plt.legend()
plt.show()
```

```python
# Handling decimals and continuous domains using numpy:
x = np.arange(0, 4.5, 0.5)
"""
- arange: array range
- Used to create a one-dimensional array with evenly spaced values within a specified range.
- Start value is inclusive; stop value is exclusive.
"""

plt.plot(x, x**2, label = 'y = x^2')
"""
- Instead of explicitly passing a list for the y-range, a function can be passed.
- This function will be applied to each value in the x-range, thereby producing the y-range.
"""

plt.legend()
plt.show()
```

```python
# Styling trick to show projection/extension:
x = np.arange(0, 4.5, 0.5)

plt.plot(x[:6], x[:6]**2, 'r', label = 'y = x^2')
plt.plot(x[5:], x[5:]**2, 'r--')
"""
- By using splice notation, we can control how many elements of the array are to be displayed.
- We can display different parts of the array in different styles.
- Use case: explaining an inference.
- Note that in the splicing notation used here, the stop value is exclusive. 
    - Thus, the stop value being six will mean that the element which corresponds to index 5 will be the last element displayed in the first format.
    - The element corresponding to index 5 will be the first element displayed in the second format.
    - Thus, there is no overlap of differently styled graphs.
"""

plt.legend()
plt.show()
```

```python
# Resizing graph:
x = np.arange(0, 4.5, 0.5)


plt.figure(figsize = (5, 3), dpi = 150)
"""
- The plt.figure() statement needs to be written BEFORE the plt.plot() statement, otherwise an error occurs.
    - It also needs to be before plt.title(), if a title is present.
- figsize:
    - Alters the aspect ratio of the graph.
    - Graph is made using figsize dimensions, taking the unit to be inches.
- DPI:
    - Alters the resolution of the graph, but not in an intuitive manner.
    - If figsize = (5, 3) and dpi = 100, the graph will be (5 * 100 = 500) by (3 * 100 = 300) pixels big.
    - When possible, keep a DPI of 300 or more, otherwise the graph may appear pixellated.
"""

plt.plot(x, x**2, label = 'y = x^2')
plt.legend()
plt.show()
```

```python
# Saving graph:
x = np.arange(0, 4.5, 0.5)

plt.plot(x, x**2, label = 'y = x^2')
plt.legend()

plt.savefig('example-of-saving-graphs.png', dpi = 200)
"""
- The plt.savefig() statement needs to be written BEFORE the plt.show() statement, otherwise the graph saved is blank.
- Note that this is different from the case of plt.figure(), which needs to be written before plt.plot() as well, and not just before plt.show()

- Used to save the graph as an image file.
- A DPI parameter can be passed to the function. This allows for different resolutions for displaying and saving the graph.
- A high DPI is recommended while saving the graph, as once again, the grpah may look choppy otherwise.
"""

plt.show()
```

### Bar Chart

```python
"""
- Takes two 1-D arrays, one for labels and the other for values, and plots a bar chart using these.
- Arrays need to be of same length, otherwise an error is thrown.

- Titles and axis labels can be added in the same mannner as basic graphs.
- Can be resized like basic graphs.
"""

labels = ['A', 'B', 'C']
values = [1, 4, 2]

plt.bar(labels, values)

plt.show()
```

```python
# Differnt kinds of hatching:
labels = ['A', 'B', 'C']
values = [1, 4, 2]

bar_chart = plt.bar(labels, values)

bar_chart[0].set_hatch('/')
bar_chart[1].set_hatch('O')
bar_chart[2].set_hatch('*')

"""
Alternative method (to prevent writing a line for each label):
hatch_patterns = ['/', 'O', '*']
for bar in bar_chart:
    bar.set_hatch(patterns.pop(0))
    
- Using pop(0) pops the first element of the list.
- If 0 were not used, the last element would be popped. This would result in a complete reversal of order of hatching.
"""

plt.show()
```

## Real World Examples

### Gas Prices Data

```python
gas = pd.read_csv('gas_prices.csv')
"""
- Creating data object by reading from CSV file.
- If file is in different folder than jupyter notebook, a relative path can be used.
"""

gas
```

```python
plt.figure(figsize = (8, 5))
# Resizing, since original scale was too small.

plt.title('Gas Prices over Time (in USD)')


plt.plot(gas.Year, gas.USA, 'r.-', label = 'USA', )
plt.plot(gas.Year, gas.Canada, 'g.-', label = 'Canada')

plt.plot(gas['Year'], gas['South Korea'], 'b.-', label = 'South Korea')

"""
- If the name of the header has a space in between, the regular '.'(dot) syntax will not work.
- The '[]' syntax needs to be used instead.
- The '[]' syntax can also be used in general.

- The shorthand notation has been used to introduce points which indicate the exact value of the graph corresponding to the xticks.
- The legend changes in accordance with the changes made to the line style.
- Order of arguments:
    - Since the shorthand notation doesn't have any keyword associated with it, it is a positional argument.
    - label, on the other hand, is a keyword argument.
    - Positional arguments cannot be placed after keyword arguments, else an error is thrown.
"""

"""
# Alternative method (to avoid writing a line for each country):
considered_countries = ['USA', 'Canada', 'South Korea']
for country in gas:
    if country in considered_countries:
        plt.plot(gas.Year, gas[country], marker = '.')
"""

plt.xticks(gas.Year[::3])
# Using ::3 to plot every third year. This is done because plotting every year makes the X-Axis very cramped.

"""
To add another year to graph (for which we don't actually have data):
plt.xticks(gas.Year[::3].tolist() + [2011])

- gas.Year is a dataframe, and cannot be manipulated directly.
- Thus, conversion to a list is required, after which the year 2011 can be added.
"""

plt.xlabel('Year')
plt.ylabel('Price in USD')

plt.legend()

plt.show()
```

#### Historgrams

```python
bins = [i for i in range(30, 101, 10)]
"""
- The 'bins' parameter defines the classes for the histogram columns.
- Obviously, it needs to be defined before plt.hist() is used.
"""

plt.hist(fifa.Overall, bins = bins, color = '#abcdef')

plt.xticks(bins)

plt.title('Distribution of Player\'s Skills in FIFA 18')
plt.xlabel('Overall Skill Level')
plt.ylabel('Number of Players')

"""
General comment:
- An issue with this visualisation is what it suggests for players with rating >= 90.
- Because so many players are present in the 60-70 range, versus the very few in the >= 90 range, it seems as though there are NO players in the latter.
- This is not actually true, as is visible from the first 5 rows of the table in the previous code block.
"""

plt.show()
```
